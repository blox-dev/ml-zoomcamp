```python
m1 = df['horsepower'].median()
mf = df['horsepower'].value_counts().index[0]
print(m1)
print(mf)

df2 = df.copy()
df2['horsepower'] = df2['horsepower'].fillna(mf)
m2 = df2['horsepower'].median()
print(m2)
print("Yes, it increased" if m2 > m1 else ("Yes, it decreased" if m2 < m1 else "No"))
```
